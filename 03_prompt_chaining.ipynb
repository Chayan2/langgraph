```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
class PromptChainState(TypedDict):
    topic : str
    outline_of_topic : str 
    blog : str
    
```

```python
def generateOutline(state : PromptChainState) -> PromptChainState:
    load_dotenv()
    model = ChatOpenAI()
    topic = state['topic']
    prompt = f"create the outline for the mentioned topic: {topic}"
    outline = model.invoke(prompt).content
    state['outline_of_topic'] = outline
    return state

```

```python
def generateBlogs(state:PromptChainState) -> PromptChainState:
    load_dotenv()
    model = ChatOpenAI()
    outline = state['outline_of_topic']
    prompt = f"generate the blog for the mentioned outline: {outline}"
    blog = model.invoke(prompt).content
    state['blog'] = blog
    return state
```

```python
#define graph 
graph = StateGraph(PromptChainState)

#define node
graph.add_node('_generate_outlines', generateOutline)
graph.add_node('_generate_blogs', generateBlogs)

# define edges
graph.add_edge(START,"_generate_outlines")
graph.add_edge("_generate_outlines", "_generate_blogs")
graph.add_edge("_generate_blogs", END)

workflow = graph.compile()
```

```python
#execute
initial_state = {
    'topic':'Blow Jobs in Sex Industry'
}

final_output = workflow.invoke(initial_state)
print(final_output)
```
